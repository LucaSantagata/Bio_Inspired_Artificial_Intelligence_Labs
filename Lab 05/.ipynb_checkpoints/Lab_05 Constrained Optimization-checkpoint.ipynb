When constraints are enforced the notion of constrained-Pareto-domination comes into play. <br> **A solution $i$ now is considered to dominate a solution $j$ if any of the following conditions are true:**

- **Solution $i$ is feasible and solution $j$ is not**
- **Solutions $i$ and $j$ are both infeasible, but solution $i$ has a smaller overall constraint violation.**
- **Solutions $i$ and $j$ are feasible and solution $i$ dominates solution $j$**

```python
class SphereCircle(Benchmark):
    def __init__(self, dimensions=2):
        Benchmark.__init__(self, dimensions)
        self.bounder = ec.Bounder([-5.12] * self.dimensions, [5.12] * self.dimensions)
        #self.bounder = ec.Bounder([-20] * self.dimensions, [20] * self.dimensions)
        self.maximize = True
    
    def generator(self, random, args):
        return [random.uniform(-5.12, 5.12) for _ in range(self.dimensions)]
        #return [random.uniform(-20, 20) for _ in range(self.dimensions)]
    
    # implements the penalty function
    def evaluator(self, candidates, args):
        fitness = []
        for c in candidates:
            f = self.f(c[0],c[1])
            if usePenalty:
                # penalty function (note that in this case we are maximizing, so penalty must be negative)
                g1 = self.g1(c[0],c[1]) # <=0
                if g1 > 0:
                    f = -1  # try to change this penalty function to handle larger search spaces
            fitness.append(f)
        return fitness
    
    def fitnessEvaluator(self, candidates, args):
        fitness = []
        for c in candidates:
            x=c[0]
            y=c[1]
            f = x**2 + y**2
            fitness.append(f)
        return fitness
    
    def constraintsEvaluator(self, candidates, args):
        constraints = []
        for c in candidates:
            g1 = self.g1(c[0],c[1]) # <=0
            constraints.append([g1])
        return constraints

    def f(self,x,y):
        return x**2 + y**2

    def g1(self,x,y):
        return x**2 + y**2 - 1
        
    def printSolution(self,c):
        f = self.f(c[0],c[1])
        g1 = self.g1(c[0],c[1])
        print("f  =", f)
        print("g1 =", g1)
        if g1 > 0:
            print("(unfeasible)")
        else:
            print("(feasible)")
```

# Adaptive penalty

The *adaptive_penalty(max_iterations, max_penalty_factor)* function performs the adaptive penalty optimization. 
It initializes the best solution and fitness with low values. It also initializes the penalty factor to 1.
It iterates for a maximum number of iterations specified by max_iterations.
In each iteration, it generates a random solution and evaluates its fitness using the *evaluate_fitness()* function with the current penalty factor.
If the obtained fitness is better than the previous best fitness, it updates the best solution and fitness.
*If the current solution violates the constraint and the penalty factor is less than max_penalty_factor, it doubles the penalty factor to increase the penalty for future iterations.*
The process continues until all iterations are completed.
Finally, it returns the best solution and its fitness.

```python
import math
import random
import matplotlib.pyplot as plt
plt.style.use('ggplot')

def sphere_function(x):
    return x[0]**2 + x[1]**2

def constraint(x):
    return x[0]**2 + x[1]**2 - 1

def evaluate_fitness(x, penalty_factor):
    constraint_value = constraint(x)
    
    if constraint_value <= 0:
        return sphere_function(x)
    else:
        penalty = penalty_factor * abs(constraint_value)
        return -penalty

def generate_random_solution():
    return [random.uniform(-1, 1), random.uniform(-1, 1)]

def adaptive_penalty_with_scatter(max_iterations, max_penalty_factor, num_solutions):
    best_solution = None
    best_fitness = float('-inf')
    penalty_factor = 1
    
    solutions = []
    fitness_values = []
    penalty = []
    
    for _ in range(max_iterations):
        solution = generate_random_solution()
        fitness = evaluate_fitness(solution, penalty_factor)
        
        if fitness > best_fitness:
            best_fitness = fitness
            best_solution = solution
        
        constraint_value = constraint(solution)
        if constraint_value > 0 and penalty_factor < max_penalty_factor:
            penalty_factor *= 2
        
        solutions.append(solution)
        fitness_values.append(fitness)
        penalty.append(penalty_factor)
    
    inside_circle = [solution for solution, fitness in zip(solutions, fitness_values) if fitness > 0]
    outside_circle = [solution for solution, fitness in zip(solutions, fitness_values) if fitness <= 0]

    inside_x = [solution[0] for solution in inside_circle]
    inside_y = [solution[1] for solution in inside_circle]
    outside_x = [solution[0] for solution in outside_circle]
    outside_y = [solution[1] for solution in outside_circle]
    
    plt.figure(figsize=(10,10))
    plt.scatter(inside_x, inside_y, color='green', label='Inside Unit Circle')
    plt.scatter(outside_x, outside_y, color='red', label='Outside Unit Circle')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Scatter Plot of Solutions')
    plt.legend()
    plt.show()

    return best_solution, best_fitness, penalty

# Set the maximum number of iterations, maximum penalty factor, and number of solutions
max_iterations = 1000
max_penalty_factor = 100
num_solutions = 100

# Perform adaptive penalty optimization and visualize the scatter plot
best_solution, best_fitness, penalty = adaptive_penalty_with_scatter(max_iterations, max_penalty_factor, num_solutions)

print("Best Solution:", best_solution)
print("Best Fitness:", best_fitness)
```

The Adaptive Penalty method progressively improves the quality of the solutions it generates by leveraging the penalty factor adjustment and the exploration-exploitation balance. Here's how the algorithm achieves this improvement over iterations:

1. **Penalty Factor Adjustment**: The penalty factor is a key component in the Adaptive Penalty method. When a candidate solution violates the constraint, the penalty factor is increased, amplifying the penalty imposed on the solution's fitness value. This adjustment encourages the algorithm to focus the search on feasible solutions that satisfy the constraint.
By increasing the penalty factor, the algorithm penalizes constraint violations more heavily, making such solutions less favorable. As a result, the algorithm gradually learns to explore and prioritize solutions that adhere to the constraint, favoring feasible regions of the search space.

2. **Exploration and Exploitation**: The Adaptive Penalty method balances exploration and exploitation during the optimization process. Initially, in the early iterations, the algorithm emphasizes exploration by generating a diverse set of candidate solutions throughout the search space. This exploration allows the algorithm to gather information about the fitness landscape and identify potentially promising regions.
As the iterations progress, the algorithm starts to exploit the knowledge gained from the exploration phase. It converges towards regions of the search space that show better fitness values, aiming to refine and improve the quality of the solutions generated.
Through this exploration-exploitation balance, the algorithm gradually learns to concentrate its search efforts on the most promising regions of the search space. It explores new areas to discover potentially better solutions while exploiting the knowledge gained to focus on more favorable regions, leading to progressive improvements in the quality of the solutions.

3. **Feedback from Fitness Evaluations**: The learning and improvement of the algorithm are facilitated by the feedback obtained from fitness evaluations. By evaluating the fitness of each candidate solution, the algorithm gains information about the quality and performance of the solutions generated.
The algorithm compares the fitness values of the candidate solutions with the best fitness obtained so far. If a candidate solution has a higher fitness value, it becomes the new best solution. This update ensures that the algorithm retains and focuses on the most promising solution found during the search, continually striving to improve the fitness and quality of the solutions generated.
Furthermore, the penalty factor adjustment, as mentioned earlier, is influenced by the constraint violation feedback obtained during fitness evaluations. By penalizing constraint violations more heavily, the algorithm learns to prioritize feasible solutions that satisfy the constraint, progressively generating higher-quality solutions.

4. **Iteration and Convergence**: The iterative nature of the Adaptive Penalty method allows the algorithm to refine the solutions and improve their quality over time. As the algorithm repeats the process of generating solutions, evaluating fitness, updating the best solution, and adapting the penalty factor, it gradually converges towards the optimal solution or a near-optimal solution.

By effectively adjusting the penalty factor, balancing exploration and exploitation, leveraging feedback from fitness evaluations, and iterating over the optimization process, the Adaptive Penalty method progressively improves the quality of the solutions it generates. It adapts its search strategy to focus on feasible regions while exploring and exploiting the search space, leading to better fitness values and improved optimization results.

# Sphere Constrained

* **Constraint Functions:** The code defines two constraint functions, **g1** and **g2**, which represent the problem constraints. These functions take the candidate solution's x and y coordinates as input and return the value of the respective constraint. 
The **g1(x, y) function represents the constraint x + y <= 4.** It ensures the circle of radius = 2. The **g2(x, y) function represents the constraint 2 - x - y <= 0**. It ensures that the difference between 2 and the sum of x and y is non-negative.

* **Penalty Function:** The evaluator() method in the SphereConstrained class includes a penalty function that handles the constraint violations and adjusts the fitness value accordingly.

After calculating the fitness value using the objective function, the code checks if the usePenalty flag is enabled. If it is, the code proceeds to calculate the penalty.
**If either g1 or g2 is greater than 0 (indicating a constraint violation), the penalty is calculated by squaring the constraint violation values and adding them together (penalty = g1^2 + g2^2)**.
The penalty is then added to the fitness value, effectively penalizing solutions that violate the constraints.

```python
class SphereConstrained(constrained_benchmarks.ConstrainedBenchmark):
    def __init__(self, dimensions=2):
        constrained_benchmarks.Benchmark.__init__(self, dimensions)
        self.bounder = ec.Bounder([-5.12] * self.dimensions, [5.12] * self.dimensions)
        self.maximize = False
        self.global_optimum = [0 for _ in range(self.dimensions)]
    
    def generator(self, random, args):
        return [random.uniform(-5.12, 5.12) for _ in range(self.dimensions)]
        
    def evaluator(self, candidates, args):
        fitness = []
        for c in candidates:
            f = self.f(c[0],c[1]) 
            usePenalty = True
            if usePenalty:
                #Change this line to handle penalty function penalty function (note that in this case we are minimizing, so we add a positive value)
                g1 = self.g1(c[0],c[1]) # <=0
                g2 = self.g2(c[0],c[1]) # <=0
                penalty = 0
                if g1 > 0 or g2 > 0:
                    penalty = g1**2 + g2**2  # Calculate penalty based on constraint violations
                f += penalty  # Add penalty to the fitness
            fitness.append(f)
        return fitness

    def constraintsEvaluator(self, candidates, args):
        constraints = []
        for c in candidates:
            g1 = self.g1(c[0], c[1])  # <= 0
            g2 = self.g2(c[0], c[1])  # <= 0
            constraints.append([g1, g2])
        return constraints

    def f(self,x,y):
        return x**2 + y**2
    
    # Implement here some constraints
    
    def g1(self, x, y):
        return x**2 + y**2 - 4  # <= 0

    def g2(self, x, y):
        return 1 - x - y  # <= 0


   
    def printSolution(self, c):
        f = self.f(c[0], c[1])
        g1 = self.g1(c[0], c[1])
        g2 = self.g2(c[0], c[1])
        print("f =", f)
        print("g1 =", g1)
        print("g2 =", g2)
        if g1 > 0 or g2 > 0:
            print("(unfeasible)")
        else:
            print("(feasible)")
```

- **What do you think is the most efficient way to handle constraints in EAs?** <br>
Handling constraints in evolutionary algorithms (EAs) can be approached in various ways, depending on the specific problem and constraints involved. Here are a few commonly used techniques for efficiently handling constraints in EAs:
1. **Penalty Functions**: One common approach is to penalize infeasible solutions by assigning a penalty to their fitness values. The penalty function increases the fitness value of infeasible solutions based on the degree of constraint violation. By incorporating penalties, the EA can still explore infeasible solutions but favors feasible solutions with lower constraint violations. This approach is relatively simple to implement but may require careful tuning of penalty values.

2. **Repair Operators**: Another approach is to use repair operators to modify infeasible solutions and bring them closer to feasibility. Repair operators adjust the values of decision variables to satisfy the constraints. These operators can be designed to repair violations while preserving the solution's overall quality. Repair operators can be applied during the variation operators, such as crossover and mutation, or as a separate step after generating offspring.

3. **Constraint Handling Techniques**: Several specialized constraint handling techniques have been developed specifically for EAs. These techniques include methods like feasibility-based selection, adaptive penalty functions, epsilon-dominance, constraint aggregation, and constraint violation ranking. Each technique has its own strengths and weaknesses, and the choice depends on the problem characteristics and specific requirements.

4. **Constraint-Handling Operators**: Constraint-handling operators explicitly consider constraints during the selection, crossover, and mutation processes. These operators bias the search towards feasible solutions and encourage the exploration of feasible regions of the search space. Examples include constrained tournament selection, constraint-aware crossover, and mutation operators that respect the feasibility of solutions.

5. **Hybrid Approaches**: Hybrid approaches combine multiple techniques to handle constraints efficiently. These approaches often integrate constraint-handling techniques with problem-specific knowledge or additional optimization methods. For example, incorporating local search or gradient-based optimization within the EA framework can help improve the search for feasible solutions.

The most efficient way to handle constraints in EAs depends on the problem characteristics, constraints, and specific requirements. There is no one-size-fits-all solution, and the choice of approach may require experimentation and careful consideration of trade-offs between exploration and exploitation, computational efficiency, and the ability to handle different types of constraints effectively.

The presence of constraints in an optimization problem does generally make the search more challenging because it restricts the feasible solution space. However, there are cases where constraints can actually make the search easier or guide it towards more favorable solutions. Here are a few scenarios where constraints can simplify or facilitate the optimization process:

1. **Narrowing the Search Space**: Constraints can effectively reduce the search space by limiting the feasible region. This reduction in the search space can make the search easier by narrowing down the possibilities and focusing the optimization process on a smaller, more relevant region.

2. **Guiding the Search Towards Desirable Solutions**: Constraints can act as guides or preferences that steer the search towards solutions that satisfy specific requirements. By explicitly defining constraints, the optimization process can prioritize finding solutions that meet the specified constraints, thereby avoiding infeasible or undesirable solutions.

3. **Modeling Real-World Constraints**: In many real-world optimization problems, constraints represent real-world limitations or requirements. By incorporating these constraints, the optimization process can ensure that the obtained solutions conform to the real-world constraints, making them more valid and practical.

4. **Reducing Uncertainty or Ambiguity**: Constraints can help to define or reduce uncertainty in the optimization problem. By specifying constraints, the optimization process can eliminate ambiguous or nonsensical solutions that do not meet the given constraints, leading to a more well-defined and meaningful search process.

5. **Exploiting Problem Structure**: Constraints can reveal structural properties of the optimization problem, providing insights into the feasible solution space. By understanding the constraints, it may be possible to exploit this structure to design more efficient search strategies or specialized algorithms that leverage the specific constraints for better exploration and exploitation.

It's important to note that while constraints can sometimes simplify the search or guide it towards desired solutions, they can also introduce additional complexity and make the search more challenging. The specific impact of constraints on the search process depends on the problem characteristics, the nature of the constraints, and the available optimization techniques and algorithms.
