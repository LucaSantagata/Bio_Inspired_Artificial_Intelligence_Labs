These rules are based on the idea that individual animals in a flock can coordinate their movements and behavior with their neighbors without any centralized control.

The three basic flocking rules are as follows:

* **Separation**: Boids should steer to avoid collisions with nearby flockmates. This is accomplished by maintaining a minimum distance from each other.

* **Alignment**: Boids should try to match the heading and speed of their neighbors. This helps the flock to move in a coordinated manner.

* **Cohesion**: Boids should steer towards the average position of their neighbors. This helps to keep the flock together and prevent individuals from becoming isolated.

In order to understand the influence of each of the three parameters, let's try the three extreme cases:

#### separation = 1, alignment = 0, cohesion =  0

Whit this parameters the Boids will exhibit strong separation behavior and avoid collisions with each other at all costs. This means that they will move away from each other as fast as possible and never come close together, even if they are part of the same flock.


```python
display.Image(r"separation.jpg", width=400, height=400)
```

#### separation = 0, alignment = 1, cohesion =  0

If the separation value is set to 0, then the objects will not attempt to maintain a minimum distance between each other. This means that the objects will not avoid collisions with each other, and may potentially overlap.

If the alignment value is set to 1, then the objects will attempt to align their velocities with their neighbors. This means that they will move in the same direction and at the same speed as their neighbors, which could result in flocking behavior.

If the cohesion value is set to 0, then the objects will not attempt to move towards the center of mass of their neighbors. This means that they will not be attracted to the group as a whole, and may move independently of each other.

```python
display.Image(r"alignment.jpg", width=400, height=400)
```

#### separation = 0, alignment = 0, cohesion =  1

If separation is set to 0, alignment is set to 0, and cohesion is set to 1 in boids, then each boid will move towards the center of mass of its local neighbors. This means that the boids will tend to cluster together, as they are attracted to each other. There will be no alignment of direction or speed among the boids, and no avoidance of other boids, so they may collide with each other. Overall, the behavior will resemble a flock of birds or a school of fish swimming in a tight group.

### Sphere 2D

```python
# -*- coding: utf-8 -*-

from pylab import *
import sys
from inspyred import ec
from inspyred import benchmarks


from utils.utils_06.inspyred_utils import NumpyRandomWrapper
import utils.utils_06.ga as ga
import utils.utils_06.es as es
import utils.utils_06.pso as pso
"""
-------------------------------------------------------------------------
Edit this part to do the exercises
"""

num_vars = 2 # Number of dimensions of the search space

args = {}

#the problem class
args["problem_class"] = benchmarks.Sphere

# other problems to try,
# see  https://pythonhosted.org/inspyred/reference.html#module-inspyred.benchmarks

# unimodal
#benchmarks.Sphere
#benchmarks.Rosenbrock

# multimodal
#benchmarks.Griewank
#benchmarks.Ackley
#benchmarks.Rastrigin
#benchmarks.Schwefel

#benchmarcks = [benchmarks.Sphere, benchmarks.Rosenbrock, benchmarks.Griewank, benchmarks.Ackley, benchmarks.Rastrigin, benchmarks.Schwefel]
#common parameters
args["max_generations"] = 100 # Number of generations
args["pop_size"] = 50 # population size

# parameters for the GA
args["gaussian_stdev"] = 0.5 # Standard deviation of the Gaussian mutations
args["mutation_rate"] = 0.5 # fraction of loci to perform mutation on
args["tournament_size"] = 2
args["num_elites"] = 1 # number of elite individuals to maintain in each gen

# parameters for the ES
args["num_offspring"] = 100 #lambda
args["sigma"] = 1.0 # default standard deviation
args["strategy_mode"] = es.INDIVIDUAL #es.GLOBAL, es.INDIVIDUAL, None
args["mixing_number"] = 1 #rho

# parameters for the PSO
args["topology"] = pso.RING #pso.RING, pso.STAR
args["neighborhood_size"] = 5   #used only for the ring topology
args["inertia"] = 0.5
args["cognitive_rate"] = 2.1
args["social_rate"] = 2.1

"""
-------------------------------------------------------------------------
"""

rng = NumpyRandomWrapper()

display = True # Plot initial and final populations
best_GA_fitnesses = []
best_ES_fitnesses = []
best_PSO_fitnesses =[]


#the problem class
args["problem_class"] = benchmarks.Sphere

# Run GA
args["fig_title"] = "GA"
best_individual_GA, best_fitness_GA, final_pop_GA = ga.run_ga(rng, num_vars=num_vars,
                                                              display=display,use_log_scale=True,
                                                              **args)
#print("Best GA fitness:", best_fitness_GA)
best_GA_fitnesses.append(best_fitness_GA)


# Run ES
args["fig_title"] = "ES"
best_individual_ES, best_fitness_ES, final_pop_ES = es.run_es(rng, num_vars=num_vars,
                                                              display=display,use_log_scale=True,
                                                              **args)
#print("Best ES fitness:", best_fitness_ES)
best_ES_fitnesses.append(best_fitness_ES)

# Run PSO
args["fig_title"] = "PSO"
best_individual_PSO, best_fitness_PSO, final_pop_PSO = pso.run_pso(rng, num_vars=num_vars,
                                                                   display=display,use_log_scale=True,
                                                                   **args)
#print("Best PSO fitness:", best_fitness_PSO)
best_PSO_fitnesses.append(best_fitness_PSO)


ioff()
show()
```

PSO tends to have a stronger focus on exploration, which means it is better suited for problems with complex or rugged search spaces where exploration is crucial. However, the 2D sphere function is a simple, convex function with only one global optimum. In such cases, GA and ES, which typically strike a balance between exploration and exploitation, can more efficiently converge towards the global optimum.
Furthermore, the 2D sphere function is a smooth and well-behaved function where gradient information is readily available. GA and ES can take advantage of gradient-based techniques or derivative information to guide their search efficiently. PSO, being a derivative-free algorithm, may struggle to utilize this information effectively, especially in cases where gradient information can significantly speed up convergence.

```python
display = False # Plot initial and final populations
best_GA_fitnesses = []
best_ES_fitnesses = []
best_PSO_fitnesses =[]


#the problem class
args["problem_class"] = benchmarks.Sphere

for _ in range(50): 
    # Run GA
    args["fig_title"] = "GA"
    best_individual_GA, best_fitness_GA, final_pop_GA = ga.run_ga(rng,num_vars=num_vars,
                                                                  display=display,use_log_scale=True,
                                                                  **args)
    #print("Best GA fitness:", best_fitness_GA)
    best_GA_fitnesses.append(best_fitness_GA)


    # Run ES
    args["fig_title"] = "ES"
    best_individual_ES, best_fitness_ES, final_pop_ES = es.run_es(rng,num_vars=num_vars,
                                                                  display=display,use_log_scale=True,
                                                                  **args)
    #print("Best ES fitness:", best_fitness_ES)
    best_ES_fitnesses.append(best_fitness_ES)

    # Run PSO
    args["fig_title"] = "PSO"
    best_individual_PSO, best_fitness_PSO, final_pop_PSO = pso.run_pso(rng,num_vars=num_vars,
                                                                       display=display,use_log_scale=True,
                                                                       **args)
    #print("Best PSO fitness:", best_fitness_PSO)
    best_PSO_fitnesses.append(best_fitness_PSO)


    ioff()
    show()
    
# Pandas dataframe
df = pd.DataFrame({"GA": best_GA_fitnesses, "ES": best_ES_fitnesses, "PSO": best_PSO_fitnesses})


sns.set_style("whitegrid")
sns.boxplot(data=df, palette="Set3", hue_order=["GA", "ES", "PSO"])
plt.legend(labels=["GA", "ES", "PSO"])
plt.ylabel("fitness values", fontsize =15)
plt.title("Best fitness 2D Sphere function comparison over 50 iterations", fontsize =15)
```

```python
import numpy as np

# Define the 2D sphere function
def sphere_function(position):
    return np.sum(position ** 2)

class Particle:
    def __init__(self, position, velocity):
        self.position = position
        self.velocity = velocity
        self.best_position = position
        self.best_fitness = float('inf')

class PSO:
    def __init__(self, num_particles, num_dimensions, max_iterations, objective_function):
        self.num_particles = num_particles
        self.num_dimensions = num_dimensions
        self.max_iterations = max_iterations
        self.objective_function = objective_function
        self.particles = []
        self.global_best_position = np.zeros(num_dimensions)
        self.global_best_fitness = float('inf')

    def initialize_particles(self):
        for _ in range(self.num_particles):
            position = np.random.uniform(-1, 1, self.num_dimensions)
            velocity = np.zeros(self.num_dimensions)
            particle = Particle(position, velocity)
            self.particles.append(particle)

    def update_particle(self, particle):
        inertia_weight = 0.2
        cognitive_weight = 1.4
        social_weight = 2

        new_velocity = (inertia_weight * particle.velocity +
                        cognitive_weight * np.random.rand() * (particle.best_position - particle.position) +
                        social_weight * np.random.rand() * (self.global_best_position - particle.position))

        particle.velocity = new_velocity
        particle.position += particle.velocity

    def optimize(self):
        self.initialize_particles()

        for _ in range(self.max_iterations):
            for particle in self.particles:
                fitness = self.objective_function(particle.position)

                if fitness < particle.best_fitness:
                    particle.best_fitness = fitness
                    particle.best_position = particle.position

                if fitness < self.global_best_fitness:
                    self.global_best_fitness = fitness
                    self.global_best_position = particle.position

            for particle in self.particles:
                self.update_particle(particle)

        return self.global_best_position, self.global_best_fitness
   

 # Create an instance of the PSO class and run the optimization for the 2D sphere function
pso = PSO(num_particles=10, num_dimensions=2, max_iterations=1000, objective_function=sphere_function)
best_position, best_fitness = pso.optimize()

print("Best position:", best_position)
print("Best fitness:", best_fitness)
```

- **When do you think it is useful to have a lower (higher) cognitive learning rate? What about the social learning rate?** <br>
The cognitive and social learning rates play a crucial role in determining how particles update their velocities and explore the search space. The learning rates control the influence of the particle's personal best position (`cognitive component`) and the best position found by neighboring particles (`social component`) during velocity updates. Adjusting these learning rates can have different effects on the algorithm's performance. Here's an overview:

1. `Lower Cognitive Learning Rate`: When the cognitive learning rate is lower, particles rely more on their personal best positions. This means that each particle places greater emphasis on its own historical best solution. It can lead to more **exploitation** of the search space, focusing on refining solutions around known good positions. A lower cognitive learning rate is useful when the search space is already well-explored, and there is a need to fine-tune solutions locally.

2. `Higher Cognitive Learning Rate`: Conversely, a higher cognitive learning rate gives more weight to the personal best position of each particle during velocity updates. It encourages more **exploration** by allowing particles to be influenced more by their own historical best solution. This can help in situations where the search space is large, complex, or has not been sufficiently explored. The higher cognitive learning rate promotes a broader exploration of the search space, increasing the chances of finding global optima.

3. `Lower Social Learning Rate`: With a lower social learning rate, particles are less influenced by the best positions found by their neighbors. **This reduces the impact of the swarm's collective knowledge and encourages particles to explore the search space independently.** A lower social learning rate is useful when the swarm encounters a deceptive or noisy objective function, where relying too much on the knowledge of other particles can lead to convergence to suboptimal solutions.

4. `Higher Social Learning Rate`: On the other hand, a higher social learning rate **increases the influence of the best positions found by neighboring particles during velocity updates.** This promotes greater information sharing and cooperation within the swarm. A higher social learning rate is beneficial in situations where the search space has few local optima and requires global exploration. It allows particles to quickly propagate promising solutions through the swarm, potentially leading to faster convergence to global optima.

In practice, finding the optimal cognitive and social learning rates often involves experimentation and depends on the specific problem being solved. Different problem domains and landscapes may require different learning rates to strike the right balance between exploration and exploitation. It is recommended to perform parameter tuning and sensitivity analysis to identify the learning rate values that yield the best performance for a given problem.


- **From a biological point of view, which neighborhood topology do you consider as the most plausible?**

From a biological point of view, the neighborhood topology that is often considered as the most plausible in the context of Particle Swarm Optimization (PSO) is the **"von Neumann" neighborhood topology.**

The von Neumann neighborhood topology is inspired by the concept of local interactions and information sharing in biological systems. It is based on the idea that each particle has a limited range of influence and interacts primarily with its nearest neighbors. In the context of PSO, the neighborhood is defined by a fixed number of particles that are considered the closest to a given particle.

The von Neumann neighborhood topology is named after the German mathematician John von Neumann and is commonly used in cellular automata and other biological models. In this topology, each particle interacts with a predefined set of neighboring particles, usually arranged in a grid-like structure. Typically, the particles are arranged in a 2D or 3D grid, and the neighbors of each particle are the particles immediately adjacent to it in the grid.

The von Neumann neighborhood topology is considered plausible from a biological standpoint because it reflects the concept of local interactions and limited information exchange. In biological systems, organisms often interact primarily with their immediate surroundings or neighbors, sharing information and adapting their behavior based on local cues. This notion of local interactions and limited range of influence is captured by the von Neumann neighborhood topology in PSO.

However, it's important to note that PSO is not directly based on biological principles, and the choice of neighborhood topology is ultimately a design decision based on the specific problem and desired properties of the optimization algorithm. While the von Neumann neighborhood topology is commonly used and considered biologically plausible, other neighborhood topologies, such as the global or ring topologies, may also be suitable depending on the problem and the desired properties of the algorithm.
