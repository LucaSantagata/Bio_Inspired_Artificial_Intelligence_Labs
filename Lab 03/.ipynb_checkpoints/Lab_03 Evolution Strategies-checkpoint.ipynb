In particular you will explore the popular ($\mu$/$\rho$,$\lambda$)-ES where $\mu$ denotes the number of parents, $\rho\leq\mu$ the number of parents involved in the producing a single offspring (mixing number), $\lambda$ the number of offspring, and *comma* selection is employed.

To start the experiments, run the next cell. Note that, as usual, you may change the seed forthe pseudo-random number generator.

Try adjusting the various parameters: $\mu$, $\lambda$, and $\rho$, (by changing, respectively, `args[pop_size]`, `args[num_offspring]`, and `args[mixing_number]` in the next cell. 
- **What happens if you make $\lambda$ smaller e.g. $\lambda=\mu$?** <br>
It means that the number of offspring or children generated in each generation is the same as the number of parents selected. With fewer offspring generated, there is reduced exploration of the search space compared to larger ùúÜ values. The algorithm focuses more on exploiting the current best solutions rather than exploring new areas. There is a lower diversity in the population. This can lead to a faster loss of diversity and potential stagnation if the initial population lacks good solutions.
 
 
- **What happens if you increase the mixing number $\rho$?** <br>
This refers to involving more parents in the recombination or mating process. This means that a higher number of parents are randomly selected from the parent population to generate each offspring.
While increasing ùúå can potentially enhance exploration and exploitation, it does not guarantee improved performance in all cases. The effectiveness of increasing ùúå depends on the characteristics of the problem, the landscape of the search space, and the balance between exploration and exploitation that is required for optimal solution discovery.
For problems with relatively simple or smooth landscapes, a smaller mixing number ùúå might already provide sufficient exploration and adding more parents in recombination may introduce unnecessary computational overhead without substantial benefits.


- **Does this confirm or contradict the conclusions you drew in the first module?**


Try out the different strategy modes (change the parameter `args[strategy_mode]`), and observe how they affect the performance of the algorithm:

- `None` means that there is no self-adaptation
- `es.GLOBAL` means each genome encodes a global step-size (mutation standard deviation)
- `es.INDIVIDUAL` means each genome encodes an independent step-size for each gene.



In the context of evolutionary strategies, the three strategy modes‚ÄîNone, es.GLOBAL, and es.INDIVIDUAL‚Äîrefer to different approaches for handling the step-size (mutation standard deviation) of the algorithm. Here's an explanation of each strategy mode:

1. **None**: In this strategy mode, there is no self-adaptation of the step-size. The mutation standard deviation remains constant throughout the evolutionary process. It does not change based on any information or feedback from the population. This means that the algorithm does not adjust the exploration-exploitation trade-off dynamically during optimization. The step-size remains fixed, and the search behavior is not adaptive.

2. es.**GLOBAL**: This strategy mode encodes a global step-size for each genome in the population. The mutation standard deviation is the same for all genes within an individual. The global step-size is adjusted based on the fitness or performance of the entire population. It allows for the exploration of the search space at a global level, as the same step-size influences all genes in all individuals. The adaptation of the global step-size aims to strike a balance between exploration and exploitation, considering the overall performance of the population.

3. es.**INDIVIDUAL**: In this strategy mode, each genome encodes an independent step-size for each gene. Each gene within an individual can have a different mutation standard deviation. The step-sizes are adjusted based on the fitness or performance of each individual separately. This mode allows for individual gene-specific adaptation, enabling fine-grained control over the exploration and exploitation trade-off. It offers flexibility to adjust the step-size based on the behavior and needs of each gene in the optimization process.

In summary, the None strategy mode keeps the step-size constant, while es.GLOBAL and es.INDIVIDUAL adapt the step-size based on the fitness or performance of the population or individuals, respectively. The choice of strategy mode depends on the problem at hand and the desired balance between exploration and exploitation during the evolutionary optimization process.

## $\lambda$ parameter analysis 

$\lambda$ value refers to an hyperparameter that determines the number of offspring to generate in each iteration of the algorithm.
In each iteration, the algorithm generates a set of offspring by mutating the current population and evaluates their fitness. The lambda parameter specifies the number of offspring to generate in each iteration.
In this section we analyzed how doeas the $\lambda$ value affects the performances of the ES

In the case of $\mu = \lambda$, this means that the number of offspring generated in each generation is equal to the number of parents selected, and the population size remains constant across generations. 
If the value of $\lambda$ is increased, this means that more offspring are generated from the selected parents. This induces a larger population size in each generation, which can be beneficial for **exploring** a larger search space. 

### $\mu = 20$,  $\lambda = [20,40,...,260]$  strategy = None

```python
display = False # Plot initial and final populations

"""
-------------------------------------------------------------------------
Edit this part to do the exercises

"""

num_vars = 10

# parameters for the ES
args = {}
args["max_generations"] = 100
args["sigma"] = 1.0 # default standard deviation

args["pop_size"] = 20 # mu
#args["num_offspring"] = 100 #lambda

args["strategy_mode"] = None
#args["strategy_mode"] = GLOBAL
#args["strategy_mode"] = INDIVIDUAL

args["mixing_number"] = 1 #rho
#args["mixing_number"] = 5

#args["problem_class"] = benchmarks.Sphere
args["problem_class"] = benchmarks.Rosenbrock
#args["problem_class"] = benchmarks.Rastrigin

"""
-------------------------------------------------------------------------
"""

args["fig_title"] = 'ES'

#seed = None


_lambdas_mu_20 = np.arange(20,280,20)
best_fitnesses_mu_20 = []
for _lambda in _lambdas_mu_20:
    args["num_offspring"] = _lambda
    
    rng = NumpyRandomWrapper(seed_)
    # Run the ES
    
    best_individual, best_fitness, final_pop = run_es(rng,num_vars=num_vars,
                                       display=display,use_log_scale=True,
                                       **args)
    # Display the results

    best_fitnesses_mu_20.append(best_fitness)
    
    
    
    if display :
        ioff()
        show()
```

```python
plt.figure(figsize = [13, 8])
plt.plot(_lambdas_mu_20, best_fitnesses_mu_20, color ="green", label = "Best fitness for differente $\lambda$ values")
plt.scatter(_lambdas_mu_20, best_fitnesses_mu_20, color = "greenyellow", s =200, edgecolor = "black")
plt.xlabel("$\lambda$ value", fontsize=20)
plt.ylabel("best fitness", fontsize=20)
plt.axhline(y=0, color='r', linestyle='--', label = "Function value at global minimum")
plt.axvline(x=20, color="grey", linestyle='--', label = "Individuals number of initial population")
plt.xticks(np.arange(20, 280, 20))
plt.title("Rosenbrock optimization strategy: None", fontsize=20, fontweight="bold")
plt.legend(loc="best",fontsize=20)
plt.show()
```

In the context of optimizing the 10D Rosenbrock function, the INDIVIDUAL strategy consistently outperforms the other strategies (None and GLOBAL) when using a larger population size ($\mu = 50$) and varying the offspring size ($\lambda$) from $50$ to $500$. There could be several reasons why the Individual strategy is the best in this scenario:

1. Adaptation to Variable Sensitivity: The Rosenbrock function is known for its sensitivity to different variables. The Individual strategy allows for gene-specific adaptation, meaning each gene can have its own step-size or mutation standard deviation. This flexibility enables the algorithm to adjust the exploration and exploitation rates individually for each variable, accommodating the varying sensitivities in the Rosenbrock function.

2. Enhanced Exploration: The Individual strategy's gene-specific adaptation promotes enhanced exploration capabilities in the search space. With a larger population size ($\mu = 50$) and a larger offspring size ($\lambda$), the algorithm has more opportunities to explore different regions and avoid getting stuck in local optima. The ability to adapt the step-size for each variable increases the algorithm's exploration potential and helps it traverse the complex Rosenbrock landscape more effectively.

3. Effective Escaping of Local Optima: The Rosenbrock function is notorious for its presence of local optima. By allowing gene-specific adaptation, the Individual strategy empowers the algorithm to escape local optima more efficiently. The algorithm can adjust the mutation standard deviation independently for each gene, enabling it to explore different regions and potentially find paths leading out of local optima.

4. Convergence Towards Global Optimum: The combination of gene-specific adaptation, larger population size, and larger offspring size likely contributes to a more effective convergence towards the global minimum of the Rosenbrock function. The Individual strategy's ability to fine-tune the step-size for each variable aids in refining the solutions and zeroing in on the global optimum.

CMA-ES (Covariance Matrix Adaptation Evolution Strategy) is a powerful optimization algorithm that excels in solving complex and high-dimensional problems. It is known for its ability to efficiently explore and converge towards optimal solutions with relatively few function evaluations and smaller population sizes compared to other evolutionary algorithms. Here are some reasons why CMA-ES is worth considering and focusing on:

1. Efficient Covariance Matrix Adaptation: CMA-ES incorporates a sophisticated mechanism for adapting the covariance matrix, which helps the algorithm effectively explore the search space. By dynamically adjusting the covariance matrix, CMA-ES can adapt its step sizes and orientations, allowing for efficient exploration of the problem landscape. This adaptive mechanism enables CMA-ES to allocate more computational effort to promising regions, leading to faster convergence and requiring fewer function evaluations.

2. Effective Handling of High-Dimensional Problems: CMA-ES is particularly well-suited for high-dimensional optimization problems. As the dimensionality increases, many optimization algorithms struggle due to the exponential growth of the search space. However, CMA-ES's adaptive covariance matrix allows it to scale well with high dimensions. By intelligently adapting the step sizes and correlations between variables, CMA-ES can navigate the complex landscape of high-dimensional problems more effectively than other algorithms, often requiring fewer function evaluations and smaller population sizes.

3. Robustness to Different Problem Types: CMA-ES demonstrates robust performance across a wide range of problem types, including both unimodal and multimodal functions. Its adaptive mechanism allows it to handle both exploration and exploitation effectively, making it capable of escaping local optima and converging towards global optima. This flexibility makes CMA-ES a versatile choice for various optimization problems without the need for problem-specific tuning.

4. Sample Efficiency: CMA-ES is known for its sample efficiency, meaning it can achieve good results with relatively few function evaluations. This can be particularly advantageous in situations where evaluating the objective function is computationally expensive or time-consuming. By achieving satisfactory results with fewer function evaluations, CMA-ES can significantly reduce the overall computational cost of the optimization process.

**- What are the advantages of self-adaptation in evolutionary computation?** <br>
As explained during the lecture, non-uniform mutation applied to continuous variables is usually done by adding some random variables from a Gaussian distribution, with zero mean and a standard deviation which controls the mutation step
size. The concept of self-adaptation represents a solution to the problem of how to adapt the step-sizes. The essential feature, and of course advantage, is that the step sizes are also included in the chromosomes and they themselves undergo variation and selection. In other words mutation step sizes are not set by the user, rather the œÉ coevolves with the solutions (the $\vec{x}$ part).
Another important advantage is the adaptation to non-stationary environments: in dynamic optimization problems, where the fitness landscape changes over time, self-adaptation can help an algorithm maintain good performance even as the environment evolves.

**- Compare the different self-adaptation strategies explored in this exercise.**<br>
In the context of evolutionary strategies, the three strategy modes‚ÄîNone, es.GLOBAL, and es.INDIVIDUAL‚Äîrefer to different approaches for handling the step-size (mutation standard deviation) of the algorithm. Here's an explanation of each strategy mode:

1. None: In this strategy mode, there is no self-adaptation of the step-size. The mutation standard deviation remains constant throughout the evolutionary process. It does not change based on any information or feedback from the population. This means that the algorithm does not adjust the exploration-exploitation trade-off dynamically during optimization. The step-size remains fixed, and the search behavior is not adaptive.

2. es.GLOBAL: This strategy mode encodes a global step-size for each genome in the population. The mutation standard deviation is the same for all genes within an individual. The global step-size is adjusted based on the fitness or performance of the entire population. It allows for the exploration of the search space at a global level, as the same step-size influences all genes in all individuals. The adaptation of the global step-size aims to strike a balance between exploration and exploitation, considering the overall performance of the population.

3. es.INDIVIDUAL: In this strategy mode, each genome encodes an independent step-size for each gene. Each gene within an individual can have a different mutation standard deviation. The step-sizes are adjusted based on the fitness or performance of each individual separately. This mode allows for individual gene-specific adaptation, enabling fine-grained control over the exploration and exploitation trade-off. It offers flexibility to adjust the step-size based on the behavior and needs of each gene in the optimization process.

In summary, the None strategy mode keeps the step-size constant, while es.GLOBAL and es.INDIVIDUAL adapt the step-size based on the fitness or performance of the population or individuals, respectively. The choice of strategy mode depends on the problem at hand and the desired balance between exploration and exploitation during the evolutionary optimization process.
